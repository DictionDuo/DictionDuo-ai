```python
import sagemaker
from sagemaker.pytorch import PyTorch
```

```python
# SageMaker 세션 및 역할
sagemaker_session = sagemaker.Session()
role = sagemaker.get_execution_role()
```

```python
# S3 경로
s3_train_uri = 's3://dd-train-bucket/train.tar.gz'
s3_output_uri = 's3://dd-conformer-bucket/model/'
```

```python
# PyTorch Estimator 생성
estimator = PyTorch(
    entry_point='train.py',
    source_dir='dictionduo_train',
    role=role,
    instance_count=1,
    instance_type='ml.g4dn.4xlarge',
    framework_version='2.2.0',
    py_version='py310',
    hyperparameters={
        'config_path': 'train_config.json'
    },
    output_path=s3_output_uri,
    base_job_name='dictionduo-conformer-train'
)
```

```python
# 훈련 작업 실행
estimator.fit({'train': s3_train_uri})
```
